```python
# Create Linear Model for this Dataset
# We can Consider that Price is a Factor Depending on Multiple Sub Factor
```

```python
X = df[['Avg. Area Income','Avg. Area House Age','Avg. Area Number of Rooms','Avg. Area Number of Bedrooms','Area Population']]
X
```

```python
y = df[['Price']]
y
```

```python
#Split in Train & Test Data set
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```python
X_test # Main Data Containing 5000 Rows and We did Test Size 30% meaning 1500 Rows
```

```python
# Lets Try to Use Linear Regression
```

```python
from sklearn.linear_model import LinearRegression
```

```python
# Need to Create An Instance for it

linearmodel = LinearRegression()
```

```python
# Main Data Alignment Has Been Destroyed , So need to Fit it Like a Good Dataset
```

```python
linearmodel.fit(X_train,y_train)
```

```python
# Need to Find The Coef_ & intercept_
```

```python
print(linearmodel.coef_)
```

```python
print(linearmodel.intercept_)
```

```python
# To Proof it Graphically
```

```python
sns.pairplot(X_test)
```

```python
print(linearmodel.coef_)
print(linearmodel.intercept_)
# Avg. Area Income , Avg Are House ,  Avg. Area Rooms, Avg. Number of Bedroom , Avg. Population
```
